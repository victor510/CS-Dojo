```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot([1,2,5], [1,5,5])
plt.title('plottings')
plt.xlabel('value')
plt.ylabel('weight')
plt.figure(figsize=(100,5))
plt.show()
```

```python
x = np.arange(1,5)
y = x**3
plt.plot([1,8,3,4], [5,7,9,10], 'go', x,y, 'r^')
plt.title('multiple sets')
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([1,2,3], [4,5,6], 'r^')
plt.title('first plot')

plt.subplot(1, 2, 2)
plt.plot(x, y, 'go')
plt.title('second plot')

plt.suptitle('my_sub plots')
plt.show()
```

```python
# bar graphs
divisions = ['DIV-A', 'DIV-B', 'DIV-C', 'DIV-D']
division_average_marks = [70, 85, 68, 98]
plt.bar(divisions, division_average_marks, color = 'green')
plt.show()
```

```python
# 3D PLOTS
from mpl_toolkits import mplot3d
ax = plt.axes(projection = '3d')
height = np.array([167,170,149,165,155,182,166,146,159,185,168,181])
weight = np.array([56,85,75,88,65,25,35,47,99,85,77,64,])
plt.xlim(140, 200)
plt.ylim(60,100)
#plt.scatter(height, weight,)
ax.scatter3D(height, weight)
ax.set_xlabel('height')
ax.set_ylabel('weight')
plt.show()
```
